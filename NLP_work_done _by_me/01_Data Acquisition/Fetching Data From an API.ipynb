```python
temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
```

```python
temp_df.head() # here you are getting only 1 page data but we saw in 
#json viewer we have 428 page so for that we will run loop to 
#extract all data
```

```python
# first we will make one empty dataframe 
df = pd.DataFrame()
df
```

```python
# for 428 page we will give 429 bcz 429 is excluisive 
for i in range(1,429):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    df = df.append(temp_df,ignore_index=True)   
```
